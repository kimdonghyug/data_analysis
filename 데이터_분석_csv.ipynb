```python
import csv
f = open('./data/seoul.csv','r',encoding='cp949') 
data =csv.reader(f,delimiter=',')
print(data)
next(data)
year = []
high = []
low=[]
for row in data:
    if row[-1] != '' and row[-2] != '':
        date = row[0].split('-')
        if 2006 <= int(date[0]):
            
            if date[1] == '09' and date[2] == '07':
                year.append(int(date[0]))
                high.append(float(row[-1]))
                low.append(float(row[-2]))
plt.title("내 생일별 기온")
plt.plot(year,high,"r",label ="high")
plt.plot(year,low,"b",label ="low")
plt.legend()
plt.show()
```

```python
f = open('./data/seoul.csv','r',encoding='cp949')
data = csv.reader(f)
next(data)
result =[]
for i in data:
    month = i[0].split('-')[1]
    if i[-1] != '':
        if month == '08':
            result.append(float(i[-1]))

plt.hist(result,bins=100,color='r')
plt.show()
```

```python
f = open('./data/seoul.csv','r',encoding='cp949')
data = csv.reader(f)
next(data)
result =[]

for i in range(31):
    result.append([])

for row in data :
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            result[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=300)
plt.boxplot(result, showfliers=False)

plt.show
```
